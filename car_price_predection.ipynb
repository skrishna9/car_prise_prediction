```python
dummies=pd.get_dummies(df['Car Model'])
```

```python
dummies
```

```python
merged=pd.concat([df,dummies],axis='columns')
```

```python
merged
```

```python
final = merged.drop(["Car Model","Mercedez Benz C class"],axis='columns')

```

```python
final
```

```python
#crteatimg a linear regression
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
x=final.drop(['Sell Price($)'],axis='columns')
```

```python
x
```

```python
y=final['Sell Price($)']
```

```python
y
```

```python
model.fit(x,y)
```

```python
#finaly i have predicted the car prise accourding to their melage ,age,and car model 
model.predict([[45000,4,0,0]])
```

```python
model.predict([[86000,7,0,0]])
```

```python
#them after i test the accuracy  and it give good accuracy 
model.score(x,y)
```
